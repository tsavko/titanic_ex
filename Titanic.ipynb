```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("data/train.csv")
```

```python
train
```

```python
train = train.drop(['Ticket', 'Cabin'], axis=1)
train = train.dropna()
```

```python
train
```

```python
# specify parameters for the graphs
fig = plt.figure(figsize=(18, 6), dpi=1600)
alpha_scatterplot = 0.2 # transparency
alpha_bar_chart = 0.55
matplotlib.style.use('ggplot')
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(2, 3)

# bar chart showing survive counts
plt.subplot(gs[0, 0])
a = train.Survived.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
plt.title('Survival Distribution')
xticks = list(a.get_xticks())
xticks[0] = 'Died'
xticks[1] = 'Survived'
a.set_xticklabels(xticks, rotation=0)

# scatterplot survival vs age
plt.subplot(gs[0, 1])
scat = plt.scatter(train.Age, train.Survived, alpha=alpha_scatterplot)
plt.xlabel('Age')
plt.grid(b=True, which='major', axis='x')
plt.title('Survival by Age')
plt.yticks([0, 1], ['Died', 'Survived'])

# classes
plt.subplot(gs[0, 2])
train.Pclass.value_counts().plot(kind='barh', alpha=alpha_bar_chart)
plt.title('Class Distribution')

# age distribution by class
plt.subplot(gs[1, :-1])
train.Age[train.Pclass == 1].plot(kind='kde')
train.Age[train.Pclass == 2].plot(kind='kde')
train.Age[train.Pclass == 3].plot(kind='kde')
plt.xlabel('Age')
plt.title('Age Distribution by Classes')
plt.legend(('1st Class', '2nd Class', '3rd Class'), loc='best')

# boarding
plt.subplot(gs[1, 2])
train.Embarked.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
plt.xticks(rotation=0)
plt.title('Passengers by Boarding Location')

gs.update(wspace = 0.3, hspace = 0.6)
```
